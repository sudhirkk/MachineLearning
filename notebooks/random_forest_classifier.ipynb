```python
# Import the libraries
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
# Set the seed, load the IRIS dataset
np.random.seed(0)
iris = load_iris()

# First split the data into training and test set by 75/25 split ratio
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25)

# Instantiate the RandomForestClassifier
rf_classifier = RandomForestClassifier(n_jobs=2, random_state=0)

# Fit the training data
rf_classifier.fit(X_train, y_train)

# Use the test data to predict
y_pred = rf_classifier.predict(X_test)

# Let's create a Pandas DataFrame from the test set and
# append the target codes, target names, predicted codes and 
# predicted names. This will help us view the actual vs predicted
df_test = pd.DataFrame(X_test, columns=iris.feature_names)
df_test['actual_codes'] = y_test
df_test['actual_names'] = pd.Categorical.from_codes(y_test, iris.target_names)
df_test['pred_codes'] = y_pred
df_test['pred_names'] = pd.Categorical.from_codes(y_pred, iris.target_names)

df_test.head(10)
```

```python
# Now we have the training and predicting completed, we can
# compute the accuracy of the model by passing X_test and y_test
# directly to the classifer.
accuracy = rf_classifier.score(X_test, y_test)
print(accuracy)
```
